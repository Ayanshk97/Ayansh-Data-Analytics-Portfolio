```python
#Iris Dataset- Exploratory Data Analysis(EDA)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load dataset (Iris from seaborn library)
df=sns.load_dataset("iris")

#show first row
df.head()
```

```python
#Data set info
df.info()

#summary statistics
df.describe().T

#check unique species
df['species'].value_counts()
```

```python
#Histogram of numerical columns
df.hist(figsize=(10,6),bins=15)
plt.show()
```

```python
#pairplot (scatterplots for all features)
sns.pairplot(df,hue="species")
plt.show()
```

```python
#correlation matrix
corr=df.corr(numeric_only=True)

#Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr,annot=True, cmap="coolwarm", fmt=".2f")
plt.title("correlation Heatmap")
plt.show()
```

```python
#Boxplots by Species
plt.figure(figsize=(10,6))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Sepal Length by Species")
plt.show()

sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Petal length by Species")
plt.show()
```

```python
#Example summary insights
print("Iris dataset has",df.shape[0],"rows and",df.shape[1],"columns.")
print("There are 3 species:",df['species'].unique())
print("Strong correlation between petal length and petal width")
print("Setosa species has distinctly smaller petal size compared to others.")
```

```python
#save processed data for future use
df.to_csv("Iris_Analyzed.csv",index=False)
print("cleaned dataset saves as Iris_Analyzed.csv")
```
